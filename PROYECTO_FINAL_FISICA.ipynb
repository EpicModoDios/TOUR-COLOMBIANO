```python
import matplotlib.pyplot as plt
import numpy as np

# Función para calcular la trayectoria del proyectil
def calcular_trayectoria(angulo, velocidad_inicial, altura_inicial, gravedad):

    """
    Calcula la trayectoria del proyectil dado un ángulo de lanzamiento, velocidad inicial,
    altura inicial y gravedad.

    Args:
    - angulo: Ángulo de lanzamiento en grados.
    - velocidad_inicial: Velocidad inicial del proyectil en cm/s.
    - altura_inicial: Altura inicial del proyectil en cm.
    - gravedad: Aceleración debido a la gravedad en cm/s^2.

    Returns:
    - posiciones_x: Posiciones en el eje X en cada instante de tiempo.
    - posiciones_y: Posiciones en el eje Y en cada instante de tiempo.
    """

    radianes = np.deg2rad(angulo)  # Convertir el ángulo a radianes
    tiempo_vuelo = (2 * velocidad_inicial * np.sin(radianes)) / gravedad  # Calcular el tiempo de vuelo
    tiempos = np.linspace(0, tiempo_vuelo, num=100)  # Crear una serie de tiempos para la simulación

    # Calcular posiciones X e Y en cada instante de tiempo
    posiciones_x = velocidad_inicial * np.cos(radianes) * tiempos
    posiciones_y = altura_inicial + (velocidad_inicial * np.sin(radianes) * tiempos) - (0.5 * gravedad * tiempos**2)

    return posiciones_x, posiciones_y



# Parámetros iniciales
velocidad_inicial = 300.875  # Velocidad inicial en cm/s
altura_inicial = 25  # Altura inicial en cm
gravedad = 980  # Aceleración debido a la gravedad en cm/s^2



# Solicitar al usuario la cantidad de ángulos a simular
num_angulos = int(input("Ingrese la cantidad de ángulos a simular: "))
angulos = []

# Solicitar al usuario los ángulos
for i in range(num_angulos):
    angulo = float(input(f"Ingrese el ángulo {i + 1} en grados: "))
    angulos.append(angulo)



# Simulaciones variando el ángulo de lanzamiento
alcances_maximos = []

for angulo in angulos:
    # Calcular la trayectoria del proyectil para cada ángulo
    trayectoria_x, trayectoria_y = calcular_trayectoria(angulo, velocidad_inicial, altura_inicial, gravedad)

    # Calcular el alcance máximo (máximo valor de la posición en X)
    alcance_maximo = np.max(trayectoria_x)
    alcances_maximos.append(alcance_maximo)

    # Graficar la trayectoria del proyectil para este ángulo
    plt.plot(trayectoria_x, trayectoria_y, label=f'Ángulo {angulo}°')

# Configuración de la gráfica
print("\n")
plt.title('Simulación del movimiento parabólico del proyectil')
plt.xlabel('Distancia en X (cm)')
plt.ylabel('Altura en Y (cm)')
plt.legend()
plt.grid(True)
plt.show()

print("\n")
print(f'Velocidad incial: {velocidad_inicial} cm/s - Altura inicial: {altura_inicial} cm')
print("\n")



# Mostrar los alcances máximos obtenidos en cada simulación
for i, angulo in enumerate(angulos):
    trayectoria_x, trayectoria_y = calcular_trayectoria(angulo, velocidad_inicial, altura_inicial, gravedad)

    # Alcance máximo horizontal
    alcance_horizontal = np.max(trayectoria_x)
    alcance_horizontal_redondeado = round(alcance_horizontal, 2)
    print(f'Ángulo {angulo}° - Alcance máximo horizontal: {alcance_horizontal_redondeado} cm')

    # Alcance máximo vertical
    tiempo_max_altura = (velocidad_inicial * np.sin(np.deg2rad(angulo))) / gravedad  # Tiempo en alcanzar altura máxima
    altura_maxima = altura_inicial + (velocidad_inicial**2 * (np.sin(np.deg2rad(angulo)))**2) / (2 * gravedad)
    altura_maxima_redondeada = round(altura_maxima, 2)
    print(f'Ángulo {angulo}° - Altura máxima alcanzada: {altura_maxima_redondeada} cm')
    print("\n")



# Generar gráfica de Posición vs Tiempo para todos los ángulos simulados
for angulo in angulos:
    trayectoria_x, trayectoria_y = calcular_trayectoria(angulo, velocidad_inicial, altura_inicial, gravedad)

    # Graficar la posición en Y en función del tiempo para cada ángulo
    plt.plot(np.linspace(0, (2 * velocidad_inicial * np.sin(np.deg2rad(angulo))) / gravedad, num=100), trayectoria_y, label=f'Ángulo {angulo}°')

# Configuración de la gráfica de Posición vs Tiempo
plt.title('Posición en Y vs Tiempo del proyectil')
plt.xlabel('Tiempo (s)')
plt.ylabel('Altura en Y (cm)')
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()
print("\n")

# Mostrar la altura máxima y el tiempo debajo de la gráfica
for i, angulo in enumerate(angulos):
    tiempo_max_altura = (velocidad_inicial * np.sin(np.deg2rad(angulo))) / gravedad  # Tiempo en alcanzar altura máxima
    altura_maxima = altura_inicial + (velocidad_inicial**2 * (np.sin(np.deg2rad(angulo)))**2) / (2 * gravedad)
    altura_maxima_redondeada = round(altura_maxima, 2)
    print(f'Ángulo {angulo}° - Altura máxima: {altura_maxima_redondeada} cm, Tiempo: {tiempo_max_altura:.2f} s')
    print("\n")


# Generar gráfica de Posición en X vs Tiempo para todos los ángulos simulados
for angulo in angulos:
    trayectoria_x, _ = calcular_trayectoria(angulo, velocidad_inicial, altura_inicial, gravedad)

    # Graficar la posición en X en función del tiempo para cada ángulo
    plt.plot(np.linspace(0, (2 * velocidad_inicial * np.sin(np.deg2rad(angulo))) / gravedad, num=100), trayectoria_x, label=f'Ángulo {angulo}°')

# Configuración de la gráfica de Posición en X vs Tiempo
plt.title('Posición en X vs Tiempo del proyectil')
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición en X (cm)')
plt.legend()
plt.grid(True)
plt.show()
print("\n")

# Mostrar los datos de posición máxima en X después de mostrar las gráficas
for angulo in angulos:
    trayectoria_x, _ = calcular_trayectoria(angulo, velocidad_inicial, altura_inicial, gravedad)
    posicion_maxima_x = np.max(trayectoria_x)
    print(f'Ángulo {angulo}° - Posición máxima en X: {posicion_maxima_x:.2f} cm')
    print("\n")




# Generar gráfica de Velocidad vs Tiempo para todos los ángulos simulados
for angulo in angulos:
    trayectoria_x, trayectoria_y = calcular_trayectoria(angulo, velocidad_inicial, altura_inicial, gravedad)

    # Calcular la velocidad en función del tiempo para cada ángulo
    velocidad_x = velocidad_inicial * np.cos(np.deg2rad(angulo))
    tiempo_total = (2 * velocidad_inicial * np.sin(np.deg2rad(angulo))) / gravedad
    velocidad_y = velocidad_inicial * np.sin(np.deg2rad(angulo)) - gravedad * np.linspace(0, tiempo_total, num=100)

    # Calcular la velocidad resultante
    velocidad_resultante = np.sqrt(velocidad_x**2 + velocidad_y**2)

    # Graficar la velocidad en función del tiempo para cada ángulo
    plt.plot(np.linspace(0, tiempo_total, num=100), velocidad_resultante, label=f'Ángulo {angulo}°')

# Configuración de la gráfica de Velocidad vs Tiempo
plt.title('Velocidad vs Tiempo del proyectil')
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad (cm/s)')
plt.legend()
plt.grid(True)
plt.show()
print("\n")

# Mostrar los datos de velocidad máxima, mínima y tiempo correspondiente después de mostrar las gráficas
for angulo in angulos:
    trayectoria_x, trayectoria_y = calcular_trayectoria(angulo, velocidad_inicial, altura_inicial, gravedad)
    velocidad_x = velocidad_inicial * np.cos(np.deg2rad(angulo))
    tiempo_total = (2 * velocidad_inicial * np.sin(np.deg2rad(angulo))) / gravedad
    velocidad_y = velocidad_inicial * np.sin(np.deg2rad(angulo)) - gravedad * np.linspace(0, tiempo_total, num=100)
    velocidad_resultante = np.sqrt(velocidad_x**2 + velocidad_y**2)

    # Encontrar el índice del tiempo medio y obtener la velocidad máxima y mínima
    indice_tiempo_medio = len(trayectoria_y) // 2
    velocidad_maxima = np.max(velocidad_resultante[:indice_tiempo_medio])  # Velocidad máxima hasta el tiempo medio
    velocidad_minima = np.min(velocidad_resultante[:indice_tiempo_medio])  # Velocidad mínima hasta el tiempo medio

    # Calcular el tiempo correspondiente a la velocidad máxima y mínima
    tiempo_velocidad_maxima = np.linspace(0, tiempo_total, num=100)[:indice_tiempo_medio][np.argmax(velocidad_resultante[:indice_tiempo_medio])]
    tiempo_velocidad_minima = np.linspace(0, tiempo_total, num=100)[:indice_tiempo_medio][np.argmin(velocidad_resultante[:indice_tiempo_medio])]

    # Imprimir la velocidad máxima, mínima y los tiempos correspondientes
    print(f'Ángulo {angulo}° - Velocidad máxima: {velocidad_maxima:.2f} cm/s, Tiempo: {tiempo_velocidad_maxima:.2f} s')
    print(f'Ángulo {angulo}° - Velocidad mínima: {velocidad_minima:.2f} cm/s, Tiempo: {tiempo_velocidad_minima:.2f} s')
    print("\n")



# Generar gráfica de Aceleración vs Tiempo para todos los ángulos simulados
for angulo in angulos:
    trayectoria_x, trayectoria_y = calcular_trayectoria(angulo, velocidad_inicial, altura_inicial, gravedad)

    # Calcular la aceleración en función del tiempo para cada ángulo
    aceleracion_y = -gravedad * np.ones(100)  # La aceleración en el eje Y es constante y es la gravedad

    # Graficar la aceleración en función del tiempo para cada ángulo
    plt.plot(np.linspace(0, (2 * velocidad_inicial * np.sin(np.deg2rad(angulo))) / gravedad, num=100), aceleracion_y, label=f'Ángulo {angulo}°')

# Configuración de la gráfica de Aceleración vs Tiempo
plt.title('Aceleración vs Tiempo del proyectil')
plt.xlabel('Tiempo (s)')
plt.ylabel('Aceleración (cm/s²)')
plt.legend()
plt.grid(True)
plt.show()
print("\n")

# Mostrar los datos de aceleración máxima después de mostrar las gráficas
for angulo in angulos:
    aceleracion_maxima = -gravedad  # La aceleración máxima en el eje Y es la gravedad
    print(f'Ángulo {angulo}° - Aceleración máxima: {aceleracion_maxima} cm/s²')
    print("\n")
```
